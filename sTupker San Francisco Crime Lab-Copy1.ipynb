## Present the Data

```python
#plotting the SF dataframe using the optional parameter
#using X and F vars which are longitutde and latitude
#to plot the graph in red and setting the marker shape to a circle using "ro".
plt.plot(SF['X'],SF['Y'], 'ro')
plt.show()
#not such a great visual so I will do some work on that
```

```python
#id the nbr of police dept districts
#build the dictionary pd_districts to associate their string to an integer
pd_districts = np.unique(SF['PdDistrict'])
pd_districts_levels = dict(zip(pd_districts, range(len(pd_districts))))
pd_districts_levels
```

```python
#using apply and lambda to add the police dept integer id
#to a new columns of the DF
SF['PdDistrictCode'] = SF['PdDistrict'].apply(lambda row: pd_districts_levels[row])
```

```python
to view the new columns
SF.head()
```

```python
#using the newly created PdDistrictCode to automaticcaly change the color
plt.scatter(SF['X'], SF['Y'], c=SF['PdDistrictCode'])
plt.show()
#this visual is so much better now 
```

```python
#now adding map packages to enhance the plot
#displayed where the crime incidents took place in SF county
#folium provides an overlay to plot this onto OpenStreet map

#for folium - the color of the marker needs to be specified using a hexadecimal value
#For this reason, the "colors" package was used and select the necessary colors
from matplotlib import colors
districts = np.unique(SF['PdDistrict'])
print(list(colors.cnames.values())[0:len(districts)])
```

```python
#now creating a color dictionary for each police district
color_dict = dict(zip(districts, list(colors.cnames.values())[0:-1:len(districts)])) 
color_dict
```

```python
#creating a map using the middle coordinates of the SF data to center the map (using "mean")
#this will reduce computation time
#"plotEvery" is used to limit amount of plotted data.
#set this value to 1 to plot all the rows (this visualize can take some time to render - depending)
map_osm = folium.Map(location = [SF['Y'].mean(), SF['X'].mean()], zoom_start = 12)
plotEvery = 50 
obs = list(zip(SF['Y'], SF['X'], SF['PdDistrict']))

for el in obs[0:-1:plotEvery]:
    
    folium.CircleMarker(el[0:2], color = color_dict[el[2]], fill_color = el[2], radius = 10).add_to(map_osm)
```
